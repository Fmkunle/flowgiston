```python
from flowgiston import *
Base = flowgiston_base(fillcolor='lightblue')
class No(Base):
    fillcolor = 'Red'
    label = 'No'
    fontcolor = 'white'
    shape = 'octagon'

class Note(Base):
    fillcolor = 'lightyellow'
    style = 'filled,dashed'
    shape = 'box'
    fontname = 'courier'

chart = FlowgistonChart(Base)

is_it_humble = chart.start("Is this a humblebrag?").edge(chart.if_("Is it humble?"))
is_it_humble.no(chart.No.node())
is_it_a_brag = is_it_humble.yes(chart.if_("Is it a brag?"))
is_it_a_brag.no(chart.No.node("By, Felicia"))
end = is_it_a_brag.yes(chart.end("Maybe")).edge(chart.Note.node("Use your judgment!"), style='dashed', dir='none')
chart
```
